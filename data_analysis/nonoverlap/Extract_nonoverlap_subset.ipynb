```python
import json
import pandas as pd
json_file = 'nonoverlap_val_YesNo.json'
source_tsv = '../../../WebQA_data_first_release/img_queries_VLP_vinvl_combinedTraining_val_beam5_img_step13_cleaned.tsv'
source = pd.read_csv(source_tsv, sep='\t')
source.head()
```

```python
from random import sample
f = open(json_file, 'r')
ids = json.load(f)
f.close()
# randomly sample 50 ids
ids = sample(ids, 50)
subset = source.loc[source['Guid'].isin(ids)]
output = {}
for index in range(len(subset)):
    temp = {}
    temp['Q'] = subset['Q'].iloc[index]
    temp['true_A'] = json.loads(subset['A'].iloc[index])[0]
    temp['pred_A'] = json.loads(subset['Output'].iloc[index])[0]
    temp['manual_label'] = 0
    output[subset['Guid'].iloc[index]] = temp

with open('VLP_vinvl_YesNo.json', 'w') as outfile:
    json_obj = json.dumps(output, indent=4)
    outfile.write(json_obj)
```

```python
# Test the accuracy
with open('VLP_vinvl_YesNo.json', 'r') as infile:
    subset = json.load(infile)
    correct = 0
    for k in subset:
        if subset[k]['manual_label'] == 1:
            correct += 1

print(correct)
```
